```python
import pandas as pd
df = pd.read_csv("http://stats.moe.gov.tw/files/detail/108/108_student.csv", thousands=',');
```

```python
df.head()
```

```python
school_names = set(df['學校名稱'])
```

```python
import json 
```

```python
with open("../dataset/ioh1500.json") as f:
    content = json.load(f)
```

```python
datas = content['data']
```

```python
depts = []
name = []
```

```python
for d in datas:
    depts.append(d['dep'])
    name.append(d['school'])
```

```python
depts = set(depts)
```

```python
name = set(name)
```

```python
school_names = name.union(school_names)
```

```python
school_names = list(school_names)
depts = list(depts)
```

```python
depts = [i for i in depts if i]
```

```python
depts = sorted(depts)
school_names = sorted(school_names)
```

```python
import collections
```

```python
school_name_dict = collections.OrderedDict()
dept_name_dict = collections.OrderedDict()
```

```python
for n in school_names:
    school_name_dict[n] = []
```

```python
for d in depts:
    dept_name_dict[d] = []
```

```python
school_name_dict
```

```python
import json
with open('school_name_dict.json', 'w', encoding='utf-8') as f:
    json.dump(school_name_dict, f, ensure_ascii=False, indent=4)
```

```python
import json
with open('dept_name_dict.json', 'w', encoding='utf-8') as f:
    json.dump(dept_name_dict, f, ensure_ascii=False, indent=4)
```
